# **Matplotlib Seaborn Visualization**

## **Use iris dataset from seaborn**

## **Create atleast 10 visuals from the data. Include Insights too**


### **violinplot, swarmplot, stripplot, boxplot, barplot, countplot,heatmap,scatter, line chart**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

## **Violinplot**

```python
plt.figure(figsize=(6, 6))
sns.violinplot(x='species', y='petal_length', data=iris,hue='species', palette='husl')
plt.title('Violin Plot of Petal Length by Species')
plt.show()

#Here, violin plot provides a detailed view of the distribution of petal lengths across different iris species.
```

## **Swarmplot**

```python
sns.swarmplot(x='species', y='sepal_length', data=iris,hue='petal_length', palette='husl')
plt.title('Swarm Plot of Sepal Length by Species')
plt.show()

#visualize the distribution of sepal length across different iris species, with an additional dimension represented by petal_length
# 'hue' parameter colors the points based on the petal_length variable.
```

## **Stripplot**

```python
plt.figure(figsize=(7, 5))
sns.stripplot(x='species', y='sepal_width', data=iris, jitter=True,hue='species', palette='Set2')   # jitter helps to make overlapping points more visible
plt.title('Strip Plot of Sepal Width by Species')
plt.show()
```

## **Boxplot**

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='species', y='petal_width', data=iris,color="red")
plt.title('Box Plot of Petal Width by Species')
plt.show()
```

## **Barplot**

```python
sns.barplot(x='species', y='sepal_length', data=iris,color="forestgreen",width=0.35,edgecolor='black')
plt.title('Bar Plot of Sepal Length by Species')
plt.show()
```

## **Countplot**

```python
sns.countplot(x='species', data=iris,hue='species', palette='viridis')
plt.title('Count Plot of Iris Species')
plt.show()
```

## **Heatmap**

```python
iris_numeric = iris[['sepal_length','sepal_width','petal_length','petal_width']]
corr = iris_numeric.corr()

plt.figure(figsize=(5, 3))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()

# 1 indicates a perfect positive correlation.
#-1 indicates a perfect negative correlation.
#0 indicates no linear correlation
```

## **Scatter plot**

```python
plt.figure(figsize=(5, 3))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris, palette='dark')
plt.title('Sepal Length  vs  Sepal Width')
plt.show()

#The scatter plot visualizes how sepal length and sepal width are related for each iris flower.
#The 'hue' parameter colors the points based on the species of the iris flowers.
#color palette used for different species is 'dark'
```

## **Line chart**

```python
sns.lineplot(x='species', y='sepal_length',data=iris, marker='*',markersize=18,color="Purple")
plt.title('Line Chart of Average Sepal Length by Species')
plt.show()
```
