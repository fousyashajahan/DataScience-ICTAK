```python
import matplotlib.pyplot as plt
freqplot = df.select_dtypes(include = ['float','int'])
freqplot.hist(figsize = (15,10))
plt.show()
```

```python
numerical_columns = ['Sale Price', 'No of Bathrooms', 'Flat Area (in Sqft)', 'Lot Area (in Sqft)','Area of the House from Basement (in Sqft)', 'Zipcode', 'Living Area after Renovation (in Sqft)']

for col in numerical_columns:
    df[col].fillna(df[col].median(), inplace=True)
df.isna().sum()
```

```python
for column in df.select_dtypes(include=['number']):
  plt.figure(figsize=(6,6))
  df.boxplot([column])
  plt.title(column)
```

```python
for column in df.select_dtypes(include=['number']):
  q1 = df[column].quantile(0.25)
  q3 = df[column].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5 * iqr
  upper_bound = q3 + 1.5 * iqr
  outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
  count = outliers.shape[0]
  print(f"Outliers in {column}: {count}")
```

```python
outlier_cols = []
for column in df.select_dtypes(include=['number']):
  q1 = df[column].quantile(0.25)
  q3 = df[column].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5 * iqr
  upper_bound = q3 + 1.5 * iqr
  outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
  count = outliers.shape[0]
  if count > 0:
    outlier_cols.append(column)
print(outlier_cols)
```

```python
for column in outlier_cols:
  q1 = df[column].quantile(0.25)
  q3 = df[column].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5 * iqr
  upper_bound = q3 + 1.5 * iqr
  df[column] = df[column].clip(lower_bound, upper_bound)
```

```python
for column in df.select_dtypes(include=['number']):
  plt.figure(figsize=(5,5))
  df.boxplot([column])
  plt.title(column)
```

```python
from sklearn.preprocessing import OrdinalEncoder
product_type_order = ['No','Yes']
OrEn = OrdinalEncoder(categories=[product_type_order])
df['Waterfront View'] = OrEn.fit_transform(df[['Waterfront View']])
df.head()
```

```python
print(df['Condition of the House'].unique())
```

```python
House_Condition = ['Bad','Okay','Fair','Good' ,'Excellent'  ]
oe = OrdinalEncoder(categories=[House_Condition])
df['Condition of the House'] = oe.fit_transform(df[['Condition of the House']])
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
for column in df.select_dtypes(include=['number']):
    mms = MinMaxScaler()
    df[column] = mms.fit_transform(df[[column]])
df.head()
```
