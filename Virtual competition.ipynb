# **Diabetes_model**

````python
diabetic_data=pd.read_csv('/content/diabetes_model.csv')
diabetic_data
````

# **Univariate Analysis**
### **Analyze individual variables (e.g., histograms, box plots).**

# ***Histogram***

````python
import seaborn as sns
sns.histplot(diabetic_data['BloodPressure'],bins=10,color='darkred',edgecolor='black')
plt.show()
````

````python
import seaborn as sns
sns.histplot(diabetic_data['Pregnancies'],bins=15,color='#614d37',edgecolor='black')
plt.show()
````

# ***Box plot***

````python
plt.figure(figsize=(5, 5))
sns.boxplot(diabetic_data['Pregnancies'],color="green")
plt.title('Box Plot')
plt.show()
````

#**Bivariate Analysis**
### **Explore relationships between pairs of variables (e.g., scatter plots, correlation matrices).**

# ***correlation matrix***

````python
corr_mat= diabetic_data[['Pregnancies','Glucose']]
corr = corr_mat.corr()
plt.figure(figsize=(6, 5))
sns.heatmap(corr, annot=True, cmap='Set2')
plt.title('Heatmap of Correlation Matrix')
plt.show()
````

# ***scatter plot***

````python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Glucose', y='DiabetesPedigreeFunction', hue='Outcome', data=diabetic_data, palette='dark')
plt.title(' Glucose vs DiabetesPedigreeFunction')
plt.show()
````

# **Multivariate Analysis**
### **Investigate interactions among multiple variables (e.g., pair plots, 3D plots)**

# ***Pair plot***

````python
plt.figure(figsize=(10, 6))
sns.pairplot(diabetic_data[['Glucose','BloodPressure','BMI']],hue='BMI',palette='viridis')
plt.show()
````

# ***3D plot***

````python
x = diabetic_data['Pregnancies']
y = diabetic_data['Glucose']
z = diabetic_data['Age']

fig = plt.figure(figsize=(15, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, alpha=0.5,color='brown')
ax.set_xlabel('Pregnancies')
ax.set_ylabel('Glucose')
ax.set_zlabel('Age')
ax.set_title('3D Scatter Plot of Glucose, BMI, and Age')

plt.show()
````
