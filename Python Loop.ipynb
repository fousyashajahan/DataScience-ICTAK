1. Take input of age of 3 people by user and determine oldest and youngest among them.

```python

a=int(input("Enter the age of first person: "))
b=int(input("Enter the age of second person: "))
c=int(input("Enter the age of third person: "))
if a>b and a>c:
    print("First person is oldest")
elif b>a and b>c:
    print("Second person is oldest")
else:
    print("Third person is oldest")
```

2. Write a program that asks the user for their exam score and calculates their grade based on the following conditions: a. A score of 90 or above is an A. b. A score between 80 and 89 is a B. c. A score between 70 and 79 is a C. d. A score below 70 is an F.

```python

score=int(input("Enter your score: "))
if score>=90:
    print("Your grade is A")
elif score>=80 and score<=89:
    print("Your grade is B")
elif score>=70 and score<=79:
    print("Your grade is C")
else:
    print("Your grade is F")
```

3. Create a program that asks the user for a list of grocery items they need to buy and organizes the list alphabetically.

```python
items = input("Enter grocery items separated by commas: ")
gro_list = [item.strip() for item in items.split(',')]
gro_list.sort()
print("\nYour grocery list sorted alphabetically:")
for item in gro_list:
    print(item)
```

5. Create a program that asks the user for the number of days they have delayed returning a library book. If the delay is less than or equal to 7 days, there is no fine. If the delay is between 8 and 30 days, apply a fine of 2 rupees per day. If the delay is more than 30 days, apply a fine of 5 rupees per day. Print the total fine amount.

```python
days=int(input("Enter the number of days you have delayed returning the library book: "))
if days<=7:
  print("No fine")
elif days>=8 and days<=30:
  print("Fine is 2 rupees per day")
  print("Total fine amount is ",2*days)
else:
  print("Fine is 5 rupees per day")
  print("Total fine amount is ",5*days)
```

7. Implement a program that calculates the factorial of a number using a loop.

```python
num=int(input("Enter a number: "))
fact=1
for i in range(1,num+1):
  fact=fact*i
print("Factorial of",num,"is",fact)
```

8. Generate the Fibonacci sequence (0, 1, 1, 2, 3, ...) up to a certain limit using a loop.

```python
limit=int(input("Enter the limit: "))
a,b=0,1
print(a)
while b<=limit:
  print(b)
  a,b=b,a+b
```

9. Create a number guessing game where the computer generates a random number and the user has to guess it within a certain number of attempts using a loop.

```python

import random
random_number = random.randint(1, 15)
max_attempts = 3
for attempt in range(max_attempts):
    guess = int(input("Guess a number between 1 and 10: "))
    if guess == random_number:
        print("Congratulations! You guessed the correct number!")
        break
    else:
        print("Wrong guess. Try again!")
if guess != random_number:
    print("Sorry, the correct number was ",random_number)
```

10. Write a menu driven calculator with following menu: 1.Addition 2.Subtraction 3.Multiplication 4.Division 5.Exit

```python
a=int(input("Enter the first number:"))
b=int(input("Enter the second number:"))
while True:
  print("\n1.Addition\n2.Subtraction\n3.Multiplication\n4.Division\n5.Exit\n")
  choice=int(input("Enter your choice:"))
  if choice==1:
   print("Result=",a+b)
  elif choice==2:
   print("Result=",a-b)
  elif choice==3:
   print("Result=",a*b)
  elif choice==4:
   print("Result=",a/b)
  elif choice==5:
   break
   print("EXIT")
  else:
   print("Invalid choice")
```
